```python
from decimal import Decimal, getcontext

MAX_DECIMALS = 1000


def e_to_n_digits(n):
    getcontext().prec = n + 20

    total = Decimal(2)  
    factorial = 1
    i = 2
    cutoff = Decimal(10) ** (-(n + 10))

    while True:
        factorial *= i
        term = Decimal(1) / Decimal(factorial)
        total += term
        if term < cutoff:
            break
        i += 1

    e_text = format(+total, "f")
    whole, _, frac = e_text.partition(".")
    frac = (frac + ("0" * n))[:n]
    if n == 0:
        return whole
    return whole + "." + frac


for n in [0, 1, 5, 15]:
    print(f"n={n}: {e_to_n_digits(n)}")


raw = input(f"Enter number of decimal places (0-{MAX_DECIMALS}): ").strip()

try:
    n = int(raw)
except ValueError:
    print("Please enter a whole number.")
else:
    if n < 0:
        print("Please enter 0 or more.")
    elif n > MAX_DECIMALS:
        print(f"Too large. Max allowed is {MAX_DECIMALS}.")
    else:
        print(f"e to {n} decimal places:")
        print(e_to_n_digits(n))
```
